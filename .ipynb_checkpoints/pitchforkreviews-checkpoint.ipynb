```python
#1) Do review scores for individual artists generally improve over time, or go down?
review_score = pd.read_sql_query(
    """
    SELECT r.pub_year,r.artist,r.score
    FROM reviews r
    INNER JOIN artists a
    ON r.reviewid = a.reviewid
    GROUP BY r.pub_year
    """,connection
)
review_score
```

```python
pub_year = review_score['pub_year'].values
score = review_score['score'].values

# Plotting the data
plt.plot(pub_year, score, marker='s', linestyle='-', color='blue')

# Adding labels and title
plt.xlabel('Publication Year')
plt.ylabel('Review Score')
plt.title('Review Score by Publication Year')

# Displaying the line chart with markers
plt.show()
```

```python
#2) How has Pitchfork's review genre selection changed over time?
generecollection = pd.read_sql_query(
    """
    SELECT r.pub_year,g.genre,COUNT(g.genre) as genre_selection
    FROM reviews r
    INNER JOIN genres g
    ON r.reviewid = g.reviewid
    WHERE g.genre <> 'None'
    GROUP BY r.pub_year,g.genre
    ORDER BY r.pub_year
    """,connection
)
generecollection.head(50)
```

```python
pivot_df = generecollection.pivot(index='pub_year', columns='genre', values='genre_selection')

# Setting the figure size
plt.rcParams['figure.figsize'] = [15, 14]

# Creating the stacked bar plot
ax = pivot_df.plot.bar(stacked=True)

# Adding bar labels
for container in ax.containers:
    ax.bar_label(container, label_type='center')

# Adding labels and title
plt.xlabel('Publication Year')
plt.ylabel('Genre Selection')
plt.title('Genre Selection by Publication Year')

# Rotating x-axis labels for better readability (optional)
plt.xticks(rotation=90)

# Displaying the stacked bar plot
plt.show()
```

```python
#3)Who are the most highly rated artists? The least highly rated artists?
rated = pd.read_sql_query(
    """
    SELECT r.pub_year,r.artist,r.score
    FROM reviews r
    INNER JOIN artists a
    ON r.reviewid = a.reviewid
    GROUP BY r.pub_year
    ORDER BY r.score DESC
    """,connection
)
rated
```

```python
rated = rated.sort_values(by='score', ascending=False)

# Plotting the bar chart
plt.rcParams['figure.figsize'] = [7, 6]  # Set figure size
plt.bar(rated['artist'], rated['score'], color="violet")
plt.xticks(rated['artist'], rated['artist'], rotation=90)  # Set x-axis ticks and labels

# Adding labels to the bars
for i, score in enumerate(rated['score']):
    plt.text(i, score, str(score), ha='center', va='bottom', fontsize=10)

# Adding labels and title
plt.xlabel('Artist')
plt.ylabel('Score')
plt.title('Score by Artist')

# Displaying the bar chart
plt.show()
```

```python
#4)Count of titles published each year
titles = pd.read_sql_query(
    """
    SELECT pub_year,COUNT(title) as no_of_titles
    FROM reviews 
    GROUP BY pub_year
    """,connection
)
titles
```

```python
plt.rcParams['figure.figsize'] = [7, 6]  # Set figure size
plt.bar(titles['pub_year'], titles['no_of_titles'])

# Adding labels to the bars
for i, num_titles in enumerate(titles['no_of_titles']):
    plt.text(titles['pub_year'].iloc[i], num_titles, str(num_titles), ha='center', va='bottom', fontsize=10)

# Adding labels and title
plt.xlabel('Publication Year')
plt.ylabel('Number of Titles')
plt.title('Number of Titles by Publication Year')

# Adjusting x-axis tick positions and labels
plt.xticks(titles['pub_year'], titles['pub_year'], rotation=90, ha='right')

# Displaying the bar chart
plt.show()
```

```python
#5)Count of authors under each type
category = pd.read_sql_query(
    """
    SELECT author_type,COUNT(author) as author_count
    FROM reviews
    WHERE author_type <> 'None'
    GROUP BY author_type
    ORDER BY author_count DESC
    """,connection
)
category.head(50)
```

```python
#6) Which titles fall under which genre
titletype = pd.read_sql_query("""
	SELECT g.genre,r.title
    FROM reviews r
    INNER JOIN genres g
    ON r.reviewid = g.reviewid
    WHERE g.genre<>'None'
    GROUP BY g.genre
""",connection)

titletype
```

```python
#To find score distribution using histogram
scoredistribution = pd.read_sql_query(
    """
    SELECT score
    FROM reviews
    """,connection
)
scoredistribution.head(10)
```

```python
plt.rcParams['figure.figsize'] = [8, 6]  # Set figure size
plt.hist(scoredistribution['score'], bins=20, edgecolor='black',color = 'brown')  # Adjust the number of bins as per your preference

# Adding labels and title
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Score Distribution Histogram')

# Displaying the histogram
plt.show()
```

```python
#genre vs review score
genrescore = pd.read_sql_query(
    """
    SELECT g.genre,r.score
	FROM reviews r
	INNER JOIN genres g
	ON r.reviewid = g.reviewid
	GROUP BY g.genre
    """,connection
)
genrescore
```

```python
average_score_by_genre = genrescore.groupby('genre')['score'].mean().reset_index()

# Plotting the bar plot
plt.rcParams['figure.figsize'] = [10, 6]  # Set figure size
plt.bar(average_score_by_genre['genre'], average_score_by_genre['score'], color='green')
plt.xlabel('Genre')
plt.ylabel('Average Review Score')
plt.title('Average Review Score by Genre')

# Rotating x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Displaying the bar plot
plt.show()
```
